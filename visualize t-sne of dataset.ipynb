```python
import os
import itertools
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
iteration_list = [250, 500]
perplexity_list = [3, 10, 30]
pca_dim_list = [25, 50]
learning_rate_list = [10, 50]
SEED=2020
```

```python
hparam_list = list(itertools.product(iteration_list, perplexity_list, pca_dim_list, learning_rate_list))
iteration, perplexity, pca_dim, learning_rate = hparam_list[0]
```

## Visualize RGB Images

```python
df = pd.read_csv(f'./csv_data/tsne_{iteration}_{perplexity}_{pca_dim}_{learning_rate}_RGB.csv')
```

```python
label2color = {'BDD':0, 'cityscape':1, 'GTA':2}
colors = [label2color[label] for label in df['img_label'].values]
```

```python
fig = go.Figure(data=go.Scattergl(
    x=df['x'].values,
    y=df['y'].values,
    mode='markers',
    marker=dict(
        color=colors,
        colorscale='Viridis',
        line_width=1
    )
))
```

```python
fig.show('png')
```
